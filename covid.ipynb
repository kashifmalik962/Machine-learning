```python
plt.figure(figsize=(15,8))
sns.heatmap(data.isnull())
```

```python
data_cols_percent = data.isnull().sum()/data.shape[0]*100
```

```python
null_cols = data_cols_percent[data_cols_percent > 20].keys()
null_cols
```

```python
data.drop(columns=['Daily_Tests', 'New_tests_per_thousand', 'Active_cases', 'Recovered',
       'Daily_recovered', 'Total_vaccinations', 'People_vaccinated',
       'People_fully_vaccinated', 'New_vaccinations',
       'Total_vaccinations_per_hundred', 'People_vaccinated_per_hundred',
       'People_fully_vaccinated_per_hundred'],inplace=True)
```

```python
data.shape
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(data.isnull())
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64'])
numerical_cols.shape
```

```python
numerical_cols.isnull().sum()
```

```python
numerical_cols.head(6)
```

```python
null_numerical_cols = [var for var in numerical_cols if numerical_cols[var].isnull().sum() > 0]
```

```python
null_numerical_cols
```

```python
plt.figure(figsize=(15,20))
for i, var in enumerate(null_numerical_cols):
  plt.subplot(4,2, i+1)
  sns.distplot(numerical_cols[var], hist=20, kde_kws={'linewidth':2, 'color':'red'}, label='Norm_Dist',)
  plt.legend()
plt.show()
```

```python
df = data.copy()
```

```python
mean_df = numerical_cols.fillna(numerical_cols.mean())
```

```python
plt.figure(figsize=(15,20))
for i,var in enumerate(null_numerical_cols):
  plt.subplot(4,2,i+1)
  sns.distplot(numerical_cols[var], hist=20, kde_kws={'linewidth':2, 'color':'red'}, label='Norm_Dist')
  sns.distplot(mean_df[var], hist=20, kde_kws={'linewidth':2, 'color':'green'}, label='Mean_Dist')
```

```python
median_df = numerical_cols.fillna(numerical_cols.median())
```

```python
plt.figure(figsize=(15,20))
for i,var in enumerate(null_numerical_cols):
  plt.subplot(4,2, i+1)
  sns.distplot(numerical_cols[var], hist=20, kde_kws={'linewidth':2, 'color':'red'}, label='Nan_Dist')
  sns.distplot(mean_df[var], hist=20, kde_kws={'linewidth':2, 'color':'green'}, label='Mean_Dist')
  sns.distplot(median_df[var], hist=20, kde_kws={'linewidth':2, 'color':'black'}, label='Medain_Dist')
  plt.legend()
plt.show()
```

```python
data.update(median_df)
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(data.isnull())
```
