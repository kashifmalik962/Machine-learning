```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/World GDP Dataset.csv')
df.head(6)
```

```python
df.head()
```

```python
df.set_index('GDP, current prices (Billions of U.S. dollars)',inplace=True)
```

```python
df.T
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = df.T
```

```python
df.head(6)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.isnull())
```

```python
df_null_percent = df.isnull().sum()/df.shape[0]*100
df_null_percent
```

```python
df_null_cols = df_null_percent[df_null_percent > 20].keys()
```

```python
df_null_cols
```

```python
df.drop(columns=df_null_cols, inplace=True)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.isnull())
```
