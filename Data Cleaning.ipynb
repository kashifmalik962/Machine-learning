```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull())
```

```python
null_val = df.isnull().sum()/df.shape[0]*100
```

```python
null_cols = null_val[null_val > 20].keys()
null_cols
```

```python
df2 = df.drop(columns=['Alley', 'FireplaceQu', 'PoolQC', 'Fence',
       'MiscFeature'])
```

```python
numerical_columns = df2.select_dtypes(include=['int64', 'float64'])
numerical_columns.isnull().sum() / df2.shape[0] * 100
```

```python
numerical_data_missing_values = [var for var in numerical_columns.columns if numerical_columns[var].isnull().sum()>0]
numerical_data_missing_values
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(numerical_columns.isnull())
```

```python
plt.figure(figsize=(15, 10))
sns.set()
for i, var in enumerate(numerical_data_missing_values):
  plt.subplot(2, 2, i+1)
  sns.distplot(numerical_columns[var], bins=20, kde_kws={'linewidth': 5, 'color': 'red'}, label="Distribution")
  plt.legend()
plt.show()
```

```python
mean_df = numerical_columns.fillna(numerical_columns.mean())
mean_df.isnull().sum()
```

```python
plt.figure(figsize=(15, 10))
sns.set()
for i, var, in enumerate(numerical_data_missing_values):
  plt.subplot(2, 2, i+1)
  sns.distplot(numerical_columns[var], bins=20, kde_kws={'linewidth': 8, 'color': 'red'}, label="Original_distribution", hist=False)
  sns.distplot(mean_df[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'}, label="Mean_distribution", hist=False)
  plt.legend()
plt.show()
```

```python
median_df = numerical_columns.fillna(numerical_columns.median())
median_df.isnull().sum().sum()
```

```python
plt.figure(figsize=(15, 10))
sns.set()
for i, var, in enumerate(numerical_data_missing_values):
  plt.subplot(2, 2, i+1)
  sns.distplot(numerical_columns[var], bins=20, kde_kws={'linewidth': 8, 'color': 'red'}, label="Original_distribution", hist=False)
  sns.distplot(median_df[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'}, label="Mean_distribution", hist=False)
  plt.legend()
plt.show()
```

```python
for i, var in enumerate(numerical_data_missing_values):
  plt.figure(figsize=(15, 10))
  plt.subplot(3, 1, 1)
  sns.boxplot(numerical_columns[var])
  plt.subplot(3, 1, 2)
  sns.boxplot(mean_df[var],orient='horizontal')
  plt.subplot(3, 1, 3)
  sns.boxplot(median_df[var])
```

```python
concatdf = pd.concat([numerical_columns[numerical_data_missing_values], 
                      mean_df[numerical_data_missing_values], 
                      median_df[numerical_data_missing_values]], axis=1)
```
