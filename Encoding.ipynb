```python
model_repl = model_count[model_count <= model_c].keys()
```

```python
manufacture_df = pd.get_dummies(df['manufacturer_name'].replace(manu_repl,'uncommon_manufactur'))
```

```python
model_name_df = pd.get_dummies(df['model_name'].replace(model_repl,'uncommon_model_name'))
```

```python
categorical_ohe_df = pd.concat([manufacture_df,model_name_df],axis=1)
```

# **Column Transform**

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv('/content/bank_marketing_dataset.csv')
```

```python
pd.set_option('display.max_columns',None)
```

```python
data.shape
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

Column tranformation

```python
transformer = ColumnTransformer(transformers=[
    ('trf1',OrdinalEncoder(categories=[['unknown','student','unemployed','housemaid','retired','admin.','self-employed',
                                         'technician','blue-collar','services','management','entrepreneur'],
                                        ['unknown','illiterate','high.school','basic.4y','basic.6y','basic.9y','university.degree','professional.course'],
                                        ['unknown','no','yes'],['unknown','no','yes'],['no','yes']]),
                                        ['job','education','default','housing','subscribed']),
      
    ('trf2',OneHotEncoder(sparse=False, drop='first',dtype=np.int32),
    ['marital','contact','month','day_of_week','poutcome'])
],remainder='passthrough')
```

```python
data.head(6)
```

```python
data['default'].value_counts()
```

```python
transform_data = transformer.fit_transform(data)
```

```python
df = data.copy()
```

```python
transform_data.shape
```

```python
data.values.shape
```

```python
data.drop(['job','marital','education','default','housing','contact','month','day_of_week','poutcome','subscribed'],axis=1,inplace=True)
```
